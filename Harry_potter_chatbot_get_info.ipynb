## load info

```python
#documents where all info is stored
documents = []
```

```python
#get info about the 7 books
docs = myloadlib.readWebsite2(["https://harrypotter.fandom.com/wiki/Harry_Potter_and_the_Philosopher%27s_Stone", "https://harrypotter.fandom.com/wiki/Harry_Potter_and_the_Chamber_of_Secrets", "https://harrypotter.fandom.com/wiki/Harry_Potter_and_the_Prisoner_of_Azkaban", "https://harrypotter.fandom.com/wiki/Harry_Potter_and_the_Goblet_of_Fire", "https://harrypotter.fandom.com/wiki/Harry_Potter_and_the_Order_of_the_Phoenix", "https://harrypotter.fandom.com/wiki/Harry_Potter_and_the_Half-Blood_Prince", "https://harrypotter.fandom.com/wiki/Harry_Potter_and_the_Deathly_Hallows"])
```

```python
#add the info to documents
documents.extend(docs)
len(documents)
```

```python
#get all the links from book pages
hyperlinks = myloadlib.loadBookURLs(["https://harrypotter.fandom.com/wiki/Harry_Potter_and_the_Philosopher%27s_Stone", "https://harrypotter.fandom.com/wiki/Harry_Potter_and_the_Chamber_of_Secrets", "https://harrypotter.fandom.com/wiki/Harry_Potter_and_the_Prisoner_of_Azkaban", "https://harrypotter.fandom.com/wiki/Harry_Potter_and_the_Goblet_of_Fire", "https://harrypotter.fandom.com/wiki/Harry_Potter_and_the_Order_of_the_Phoenix", "https://harrypotter.fandom.com/wiki/Harry_Potter_and_the_Half-Blood_Prince", "https://harrypotter.fandom.com/wiki/Harry_Potter_and_the_Deathly_Hallows"])
```

```python
#Takes around 20 minuts
#before sublinks 854
start = time.time()
#get info from all the links
for link in hyperlinks:
    print(link)
    docs = myloadlib.readWebsite2("https://harrypotter.fandom.com" + link)
    documents.extend(docs)
    sublinks = myloadlib.loadURL("https://harrypotter.fandom.com" + link)
    for sublink in sublinks:
        if sublink not in hyperlinks:
            hyperlinks.append(sublink)
end = time.time()
print((end - start)/60)
```
